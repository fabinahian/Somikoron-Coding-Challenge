# üîñ Tokenizing the Words

````python
#tokenization of words

tokenized_words = bn.word_tokenize(paragraph)
print("words = ",tokenized_words)
print("\n")

#length of the tokenized words

tokenized_words_length = len(tokenized_words)
print("Length = ", tokenized_words_length)
print("\n")
````

# üîñ Tokenizing the Sentences

````python
#tokenization of sentences

tokenized_sentences = bn.sent_tokenize(paragraph)
print("sentences = ",tokenized_sentences)
print("\n")

#length of the tokenized sentences

tokenized_sentences_length = len(tokenized_sentences)
print("Length = ", tokenized_sentences_length)
print("\n")
````

# üîñ Stemming vs Lemmatization

Both Stemming and Lemmatization share the same goal, which is to bring words to their base forms.
In everyday use, words take on multiple forms according to the demands of Grammar. For example, 
> I eat, he eats, she has eaten - these three sentences represent different faces of the word 'eat'.

While both Stemming and Lemmatization tries to find the base form of words, they do the task differently. Here are some key differences between the two of them:

**Stemming**

1. Stemming works by removing the last few characters of words without any regard to context.
2. It has a higher error rate since it returns whatever is left after the reduction of affixes, even if the remaining bit isn't a word.
3. Example: 
        a. walking --> walk
        b. caring --> car
        c. stripes --> strip
        
**Lemmatization**

1. Lemmatization works by taking contexts into consideration and using a dictionary to get the base form of words.
2. It has a lower error rate because it always returns dictionary words as output.
3. Example:
        a. walking --> walk
        b. caring --> care
        c. stipes (noun) --> stripe
           stripes (verb) --> strip
